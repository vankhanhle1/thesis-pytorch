```python
df = pd.read_csv(path + 'test-df.csv')
terp_ref = []
terp_hyp = []
for i in range(len(df)):
    s0, s1 = remove_special_characters(df.iloc[i, 0]), remove_special_characters(
                df.iloc[i, 1])
    annotation = ' ([kle][msrp][{index}])' 
    terp_ref.append(s0 + annotation.format(index=i*2)) 
    terp_hyp.append(s1 + annotation.format(index=i*2))
    terp_ref.append(s1 + annotation.format(index=i*2+1))
    terp_hyp.append(s0 + annotation.format(index=i*2+1))
```
